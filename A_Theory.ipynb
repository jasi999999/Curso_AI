## Unit 1. Numpy Array Data Structure for Optimal Computational Performance

### Unit 1. 0. Theory

- Memory is **expensive**: Numpy is implemented in C for maximum **optimization**. Depending on the algorithm there are **memory** or **speed** costs.
- **Data structure**: data finding, organization, management and storage are **very important**.
- **Datatypes**: primitive, linear, non-linear and file .
- **Random access in Array**: using the index, access to *wherever* you want while keeping the **array structure**, `O(1)`.
- **Adding** or **deleting in Array**: is not that easy and can have increased cost `O(∏ᵢ₌₁ⁿ dimᵢ)`. **Adding at the end** costs `O(1)`.
- - Data and algorithms: **finding data** is important, it can be speedy but at a **memory cost**. In addition, **allocating data** increases the difficulty of the problem.  
- Library **Numpy**: implements **multidimensional array objects**, **matrixes**, ... Supports **discrete Fourier transforms**, **basic algebra, statistics, simulations...**.It is implemented in C, uses **element-by-element operations**.  
- **Array DataType**: created as `np.array([n, n, ...])`, allocates memory, "learns" that it is an *Integer* array. Identical objects will have different locations in memory. Comparing two-array objects returns a new array with a comparison of every single object one by one; because of **element-by-element operations**. `ndarray` means `n-dimensional array`.`32-bit integers`. Do **not use dictionaries or sets inside an array**. Use array to implement lists made of lists as matrixes. Runs through a matrix through rows. It has **type inferring** which works very well to **prevent memory waste**. They have **fixed sizes**, so do the inside elements.
- **Vectorization**: applies the same `operation` to a big chunk of data, treating it as a **block of data**. Maximizes efficiency in orders of magnitude.  
- **Copying an array**: the way in which the copy will be **allocated in memory** can be defined (columns, rows, ...). **Making copies can use more memory and cost more than referencing or viewing the same array**. So, only use copy when you need to modify the previous object and still retain the original. Do **not use copies when you are not modifying** the object.  
- **Memory usage**: if there is a string array, when a long list made of small words has a very long word, every *box* will be as big as the longest word, therefore **wasting memory space**. Also the **data type can create waste (such as saving numbers as strings)**. It takes a lot to do a `reshape` than to `create` an array.

Be careful also using `ndarray.view(encoding)` function. **Matrix manipulation**: **WILL PROBABLY AFFECT MEMORY ALLOCATION**. In addition, when deleting a big matrix, it might be interesting to create new matrices of the smaller matrices that share memory with it, as **MEMORY WILL NOT BE FREED UNTIL THE WHOLE CHUNK IS COMPLETELY FREE** (check `np.copy()`).

The axis can be defined, so that addition is done to `rows (0)` or `columns (1)`. As the creation of an array is first rows and then columns, adding a new row is just appending an array at the end, whereas adding a column means adding **something to every column (i. e. `[[3],[9]` appended to a `2 x 2` matrix)**.  
- Function `np.delete(array, element, axis)`: deletes an element of the array. **Careful: Incorrectly modifying a matrix may result in an array return**.

- Function `np.sqrt(array)`: returns the square root of the given numbers.
- Function `np.random.seed(n)`: establishes the seed of the semi-random number generator.

### 2. 1. Series

- Library **Pandas**: tabular data manipulation library (most extended).
- Check the **Pandas Documentation**:
  - [[General Functions]](https://pandas.pydata.org/docs/reference/general_functions.html).
  - [[Dataframe Functions]](https://pandas.pydata.org/docs/reference/frame.html).
- `Series`: similar to Numpy array (`1 datatype` and `1 dimension`), with added `title`.
- `DataFrame`: a group of *series*, with different datatypes.
- Parameter `inplace=Bool`: to modify the original.
- Function `pd.Series(list)`: creates a pandas series from iterable data, but **also dictionaries** (where indexes are the keys of the dictionary).
- Function `ser.index`: returns a **compressed version of the series index (WHENEVER IT CAN BEST REDUCE IT, such as ranges)**.
- Function `ser.values`: returns the values of the series.
- Function `ser.dtype`: returns the datatype property.
- Function `pd.DataFrame(data, index, columns, dtype)`: creates a data frame with the information of the given object, a list can be passed to describe the index or columns names.
- Library `from IPython.display import display`: to be able to see several data frames at the same time.
- Function `display(dataframe)`: displays a data frame.
- Function `df.columns`: returns the `index` of the columns, similar to `ser.index`.
- Function `ser.name`: returns the name of the series, can be modified.
- Function `ser.sort_values`: returns a series sorted by values.
- Function `ser.sort_index`: returns a series sorted by index values.
- Function `ser.unique()`: returns the unique values of a series.
- Function `ser.nunique()`: returns the number of unique values.
- Function `ser.value_counts()`: counts all the equal values.
- `Pandas operations`: are made **to values, element-to-element** and involve all classical arithmetical operations.
- Functions `ser.sum()`, `ser.mean()`, `ser.median()`: return a value of the `sum`, `mean` and `median` of all series values.
- Function `ser.apply(lambda function)`: applies a lambda function to the series.

```python
import pandas as pd
import numpy as np
from IPython.display import display

print(pd.Series(['Male', 'Female', 'Male', 'Male', 'Female']))
ser = pd.Series({'a': 1, 'b': 2, 'c': 3})
print(ser)
print(ser.index)                                         # Returns a reduced version of thseries index (whenever it can)
ser2 = pd.Series(np.arange(30))
print(ser2.index, end="\n\n")                            # Returns the compressed version of the index

list_data = ['2019-08-10', 3.14, 'ABC', 100, True]
ser3 = pd.Series(list_data)                              
print(ser3, end="\n\n")

d = {
    'Name': ['Brown', 'Henry', 'Elizabeth'],
    'Age': [22, 35, 58],
    'Sex': ['Male', 'Male', 'Female']
}
df1 = pd.DataFrame(d)
display(df1)
print(df1['Age'])                                        # Access the series of a data frame
print(df1['Age'][0])                                     # Access an element of the series of a data frame
# Indexes can be manipulated  but with all-dictionaries inside the dictionary
d2 = {
    'Name': {0: 'Brown', 1: 'Henry', 2: 'Elizabeth'},
    'Age': {0: 22, 2: 35, 1: 58},
    'Sex': {0: 'Male', 2: 'Male', 1: 'Female'}
}
df2 = pd.DataFrame(d2)
display(df2)

df3 = pd.DataFrame(np.arange(1, 10).reshape(3, 3))
display(df3)

my_data = [220, 13, 23, 34, 234]
ser5 = pd.Series(my_data, index=['A', 'B', 'A', 'A', 'C'])
ser5.sort_values(inplace=True)
display(pd.DataFrame(ser5))

ser6 = pd.Series(np.arange(0, 50, step=10), index=['a', 'b', 'c', 'd', 'e'])
print(ser6)
print(ser6['a':'d'], end="\n\n")                         # Slicing works fine, but includes the last one

ser_height = pd.Series([160, 170, 180], name="height")
print(ser_height.apply(lambda x: x + 10), end="\n\n")
```

### 2. 2. DataFrames

- Function `pd.read_csv('name.csv')`: reads the data from a given `csv` file.  
- Function `df.info()`: returns a brief summary of the given Dataframe.  
- Function `df.head(n)`, `df.tail(n)`: returns the first or last `n` rows of a Dataframe.  
- Function `df[[col1_name, coln_name]]`: accesses the values of several given columns, if only access to one column, no need of double brackets `[[]]`.  
- Function `df.columns = [list]`, `df.index = [list]`: the **new given list will be the new index or column names**.  
- Function `df.rename(col={}, index={}, inplace=bool)`: returns or modifies the values of the dictionary inside index or columns.  
- Function `df.drop(labels, axis=dim, inplace=bool)`: drops the chosen columns or rows, a list can be passed for several options.  
- Function `df.loc[index]`: returns the selected row `or rows ([[]])` by given index. **ROW SELECTION**.  
- Function `df.iloc[[num1, num2]]`: returns the selected row `or rows ([[]])` by given index numbers. **ROW SELECTION**.  
- Function `df.loc[dim1, dim2]`, `df.iloc[dim1, dim2]`: returns the data allocated in that point. Example: `df.iloc[0, [2, 3]]`.  
- Function `df = df.loc[:, cols_order]`: **reorders the data frame columns and its data** (actually creates a new Dataframe).  
- `df[col]`: returns a pandas Series.  
- `df[[col1, col2]]`: returns a pandas DataFrame.  
- Function `df['new'] = val`: **column addition**. Can be done using one value for all or a list with all designated values (example is one value for all). **This action is INPLACE**. **Column addition has a LOW COST**.  
- Function `df.loc[out_of_range] = [val]`: row addition **only if the index DOES NOT EXIST (not reliable, better use `pd.concat()`)**. Can be done using one value for all or a list with all designated values (example applies a list). **Row addition has a HIGHER COST, AS IT CREATES A NEW DATAFRAME**.  
- Function `df.reset_index(inplace=bool)`: returns the index column as column `0` or drops it. Used to **manipulate index names**.  
- Function `df.set_index([col], inplace=bool)`: destroys the current index and places the new one **or ones (if a list is given)**. **Beware of losing the current index**.  
- Function `df.transpose()`, `df.T`: transposes the given data frame.  
- Function `df.reindex(new_index, fill_value=n)`: returns a new data frame and fills every empty spot with `NaN` or the given value.  
- Function `df.sort_index(ascending=bool, inplace=bool)`: **reorders the data frame**, can be applied to rows or columns.  
- Pandas and DataFrame `operations`: element-to-element typical operations. **Operating with NaN or empty numbers will return a NaN**.  
- Function `ser1.op(ser2, fill_value=n)`: operations like `add`, `sub`, `mul`, `div` can be performed in this way to fill with numbers.  
- Function `pd.merge(df1, df2, on=common_col_name, how)`: **combines Dataframes using things in common** using the `common_col_name` (`left_on` and `right_on` can be used to select what columns we try to combine) and the `how` decision (`inner`: **intersection**, `left`: takes **left and fills**, `right`: takes **right and fills**, `outer`: **fills everything**). Looks for similar columns and merges them. If no common column, an error message will appear.  
- Function `pd.concat([df1, df2], sort=bool, axis=dim)`: puts one data frame **next to each other**. **ALIGNS WITH THE INDEX** and **fills with NaN when labels don't match**.
- Function `df.sort_values(by=['chosen_col1', 'chosen_col2'])`: **sorts** the values by the chosen column or columns (when there are coincidences).  
- Function `pd.MultiIndex.from_tuples(list_of_tuples)`: using a `zip()` function we make a list of combinations of indexes to create a **multi-index TO BE USED LATER**. How to:
  - **Make outer and inner indexes**: use lists to create the indexes.
  - **Combine the indexes**: use a `list(zip(outer, inner))` to make a tuple list.

- Check the **MatPlotLib Documentation**:

- Function `plt.show()`: displays the plot, **DRAWS OVER PREVIOUSLY DRAWS**.
- Function `plt.bar(x, y, color)`: builds a **barchart plot**.
- Function `plt.hist(x, bins, color, density=bool, rwidth=float, alpha=float)`: builds an **histogram plot**. Receives the data, divides it into the bins,  and `density`set as True gives the **relative probability instead of raw counts**. Different coloured plots can be plotted over one another.
  - `density`: normalizes the histogram. **Area must sum 1**, giving relative probability.
  - `rwidth`: sets the width between bars.
  - `alpha`: sets the transparency of the plot.
- Function `plt.boxplot([data1, d2, d3, ...], vert=bool)`: builds one or several boxplots, with `vert` used to set the axis.

- Function **`plt.suplot(nrows, ncols, figsize, ...)`**: **REALLY POWERFUL WHEN DRAWING SEVERAL PLOTS AT A TIME**.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = ['a', 'b', 'c']
y = np.random.randint(0, 10_000, size=3)
plt.bar(x, y, color='purple')
plt.title('Bar')
plt.show()

# Figure creation
fig0, axes0 = plt.subplots(nrows=2, ncols=3, figsize=(6, 6))

# Plot in position 0, 0
axes0[0, 0].hist(np.random.randn(10_000), bins=30)

# Plot in posiition 0, 1
x01 = np.linspace(0, 10, 100)
y01 = np.sin(x01)
axes0[0, 1].plot(x01, y01, linewidth=2, linestyle=':')
axes0[0, 1].plot(x01, np.cos(x01), linewidth=2, linestyle='--')
axes0[0, 1].set_title('Line plot')
axes0[0, 1].legend(['sin(x)', 'cos(x)'])

# Plot in position 0, 2
x02 = np.random.randn(10_000)*10 + 3
y02 = np.random.randn(10_000)*10 + 5
z02 = np.random.randn(10_000)*10 + 1
axes0[0, 2].boxplot([x02, y02, z02], 0)
axes0[0, 2].set_title('Boxplot')

# Plot in position 1, 0
x10 = np.random.randn(10_000)
y10 = np.random.randn(10_000) + 1
axes0[1, 0].hist(x10, bins=10, density=True, color='green', rwidth=0.95, alpha=0.5)
axes0[1, 0].hist(y10, bins=10, density=True, color='purple', rwidth=0.95, alpha=0.5)
axes0[1, 0].set_title('Histogram')

# General figure instructions
plt.tight_layout()
plt.show()
```
